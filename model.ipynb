```python
import numpy as np 
import pandas as pd
from PIL import Image #pillow library
```

```python
import tensorflow as tf 
from keras.utils import to_categorical
from keras import models, layers
import sys, os
import json
import glob
```

```python
with open('C:\\Users\\root\\Desktop\\printed_data\\printed_data_info.json','r',encoding='utf-8') as f:
    Korean_data_print =json.load(f)

#test= glob.glob('C:\\Users\\root\\Desktop\\printed_data\\syllable\\*.png')

print(len(Korean_data_print))
```

```python
print(Korean_data_print.keys(),len(Korean_data_print['annotations']))
```

```python
count = 0
for i,Data in enumerate(Korean_data_print['annotations']):
    if Data['id'] != Data['image_id']:
        count+=1

print(count) #id가 곧 image_id ok
```

```python
#사용할 
Korean_id = []
Korean_text = []

for i,Data in enumerate(Korean_data_print['annotations']):
    if Data['attributes']['type']== '글자(음절)':
        Korean_id.extend([Data['id']])
        Korean_text.extend([Data['text']])

print(len(Korean_id),len(Korean_text))

#메모리 확보를 위해 정리.
del(Korean_data_print)

```

```python
#data 파일에 현제 치대는 532659가지임
Korean_train_id = Korean_id[:400000]
Korean_train_text = Korean_text[:400000]

Korean_test_id = Korean_id[400000:532659]
Koreann_test_text = Korean_text[400000:532659]

syllable = list(set(Korean_text)) #집합으로 먼저 만들고 리스트화
print(len(Korean_train_id),len(Korean_test_id))
print("음절 총 개수 : ",len(syllable))
```

```python
#메모리 확보를 위해 file Data 정리
del(Korean_id)
del(Korean_text)
```

```python
#train과 test id의 image를 담을 공간 생성
x_train = np.zeros((400000,32,32,3),'uint8') 
x_test = np.zeros((132659,32,32,3),'uint8')
```

```python
for i, ID in enumerate(Korean_train_id):
    if i%1000 ==0:
        print(i,"번째입니다")
    Image_addr = 'C:\\Users\\root\\Desktop\\printed_data\\syllable\\'+str(ID)+'.png'
    Korean_Image = Image.open(Image_addr)
    Korean_Image = Korean_Image.resize((32,32))
    Korean_Image_Array =np.array(Korean_Image,'uint8')
    x_train[i] =Korean_Image_Array
```

```python
for i,ID in enumerate(Korean_test_id):
    if i%1000 ==0:
        print(i,"번째입니다")
    Image_addr = 'C:\\Users\\root\\Desktop\\printed_data\\syllable\\'+str(ID)+'.png'
    Korean_Image = Image.open(Image_addr)
    
    Korean_Image = Korean_Image.resize((32,32))
    Korean_Image_Array = np.array(Korean_Image,'uint8')
    x_test[i] = Korean_Image_Array
```

```python
x_train = x_train.astype('float32')/255.0
x_test = x_test.astype('float32')/255.0

del(Korean_train_id)
del(Korean_test_id)

print(len(x_train),len(x_test))
```

```python
#각 음절 Indexing
label_train = []
label_test = []

syllable_to_index = {syllable: index for index, syllable in enumerate(syllable)} 

for syllables in Korean_train_text:
    if syllable_to_index.get(syllables) is not None: #get은 key기반 if not key return 'none'
        label_train.extend([syllable_to_index[syllables]])

for syllables in Koreann_test_text:
    if syllable_to_index.get(syllables) is not None:
        label_test.extend([syllable_to_index[syllables]])


del(Korean_train_text)
del(Koreann_test_text)
print(len(label_train),len(label_test))
```

```python
label_train =to_categorical(label_train)
label_test = to_categorical(label_test)
len(label_test[0])
```

```python
##CNN layer 추가 후 Parameter 설정
#model.add(Conv2D(input_shape = (X_train.shape[1], X_train.shape[2], X_train.shape[3]), filters = 50, kernel_size = (3,3), strides = (1,1), padding = 'same', kernel_initializer='he_normal'))
CNN = models.Sequential()
CNN.add(layers.Conv2D(128, (3,3), activation='relu',input_shape=(32,32,3)))# 왜 He 초기값은 그런거냐...
CNN.add(layers.MaxPooling2D((2,2)))

CNN.add(layers.Conv2D(256, (3,3), activation='relu'))
CNN.add(layers.MaxPooling2D((2,2)))

CNN.add(layers.Conv2D(512, (3,3), activation='relu'))
CNN.add(layers.Flatten())

CNN.add(layers.Dense(512, activation='relu'))

CNN.add(layers.Dense(11172, activation='softmax'))
CNN.summary()
```

```python
#CNN 함수 설정 후 Training
from keras import optimizers
CNN.compile(optimizer = optimizers.RMSprop(lr=0.001),
loss = 'categorical_crossentropy',
metrics = ['accuracy'])


hist = CNN.fit(x_train, label_train,
epochs = 40, batch_size = 200,                                 
validation_split=0.2)
```

```python
test_loss, test_acc = CNN.evaluate(x_test, label_test)
print('test_loss:      ',test_loss)
print('test_accuracy:  ',test_acc)
```
